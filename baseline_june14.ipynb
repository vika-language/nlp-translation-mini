```python
# 预定义评价维度、正负情感词典
positive_words = {"good", "great", "nice", "excellent", "cheap"}
negative_words = {"terrible", "bad", "awful", "poor", "expensive"}
aspect_words = {"service", "price", "food", "environment", "taste"}

sentence = "The service is terrible but the price is good."
lower_sent = sentence.lower()
words = lower_sent.split()

quadruple_results = []

# 遍历匹配，生成【评价目标、观点词、情感极性、类别】四元组
for aspect in aspect_words:
    if aspect in words:
        # 匹配正面情感
        for opinion in positive_words:
            if opinion in words:
                quadruple_results.append({
                    "评价目标(aspect)": aspect,
                    "观点词(opinion)": opinion,
                    "情感极性(sentiment)": "正面(positive)",
                    "类别(category)": "服务价格类"
                })
        # 匹配负面情感
        for opinion in negative_words:
            if opinion in words:
                quadruple_results.append({
                    "评价目标(aspect)": aspect,
                    "观点词(opinion)": opinion,
                    "情感极性(sentiment)": "负面(negative)",
                    "类别(category)": "服务价格类"
                })

# 格式化打印四元组结果
print("=== 属性情感四元组抽取结果 ===")
for idx, item in enumerate(quadruple_results, 1):
    print(f"第{idx}组：{item}")
```

```python
# 定义情感词典（正面/负面）
positive_words = ['good', 'great', 'excellent', 'wonderful', 'nice', 'beautiful', 'perfect', 'love', 'like', 'amazing']
negative_words = ['terrible', 'bad', 'awful', 'poor', 'horrible', 'disgusting', 'hate', 'worse', 'ugly']

# 定义方面词及其类别映射
aspect_category = {
    'service': '服务类',
    'price': '价格类',
    'hill': '意象类',
    'bird': '意象类',
    'path': '意象类',
    'fisherman': '人物类',
    'boat': '物象类',
    'snow': '物象类'
}

def extract_quadruples(text):
    words = text.lower().split()
    quadruples = []
    for aspect, category in aspect_category.items():
        if aspect not in words:
            continue
        # 查找方面词附近的观点词（这里简单取后面第1个词作为观点词）
        idx = words.index(aspect)
        if idx + 1 < len(words):
            opinion = words[idx + 1].strip('.,;!?')
            if opinion in positive_words:
                sentiment = 'positive'
            elif opinion in negative_words:
                sentiment = 'negative'
            else:
                sentiment = 'neutral'
            quadruples.append((aspect, opinion, sentiment, category))
    return quadruples

# 定义《江雪》许渊冲和斯奈德的译文
xu_text = "From hill to hill no bird in flight; From path to path no man in sight. A lonely fisherman afloat, Is fishing snow in lonely boat."
snyder_text = "These thousand peaks cut off the flight of birds, On all the trails, human tracks are gone. A single boat — coat — hat — an old man! Alone fishing chill river snow."

print("=== 许渊冲译文四元组提取结果 ===")
result_xu = extract_quadruples(xu_text)
if result_xu:
    for quad in result_xu:
        print(f"方面: {quad[0]}, 观点: {quad[1]}, 情感: {quad[2]}, 类别: {quad[3]}")
else:
    print("未提取到任何四元组（可能是因为诗歌文本缺乏明确的'方面词+观点词'结构）")

print("\n=== 斯奈德译文四元组提取结果 ===")
result_snyder = extract_quadruples(snyder_text)
if result_snyder:
    for quad in result_snyder:
        print(f"方面: {quad[0]}, 观点: {quad[1]}, 情感: {quad[2]}, 类别: {quad[3]}")
else:
    print("未提取到任何四元组（诗歌语言中方面词与观点词没有明显共现）")
```

由于诗歌语言高度凝练，极少出现“方面词 + 明确的观点形容词”这类显式结构（如“service is terrible”），因此基于词典和共现规则的简单方法无法抽取出四元组。

这说明：

- 对于文学翻译研究，需要更复杂的模型（如序列标注、预训练模型微调）才能捕捉隐含的评价。
